#### Chart - 1 : 
Q1. Are Shipments Managed by specific teams( Eg:- PMO-US) more likely to be delivered on time compared to others?

````python
# Chart - 1 visualization code
# Calculate delivery delay in days
fd['Delivery Delay (days)'] = (fd['Delivered to Client Date'] - fd['Scheduled Delivery Date']).dt.days

# Create a flag for on-time delivery (delivered on or before scheduled date)
fd['On Time Delivery'] = fd['Delivery Delay (days)'] <= 0
````

````python
fd.head(5)
````

````python
# Group by 'Managed By' to analyze on-time delivery performance
on_time_by_manager = fd.groupby('Managed By')['On Time Delivery'].mean().sort_values(ascending=False)
````

````python
# Plot the bar chart
plt.figure(figsize=(10, 6))
on_time_by_manager.plot(kind='bar', color='blue', edgecolor='black')
plt.title('On-Time Delivery Rate by Shipment Manager')
plt.ylabel('On-Time Delivery Rate')
plt.ylim(0, 1.1)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add percentage labels above each bar
for i, val in enumerate(on_time_by_manager):
    plt.text(i, val + 0.02, f"{val*100:.1f}%", ha='center', va='bottom', fontsize=10, fontweight='bold')
    
plt.tight_layout()
plt.show()
````

This chart allows to easily compare on-time delivery performance across shipment managers/regional offices. It helps to tell who's doing well & who's lagging behind. It highlights performance differences.

#### Chart - 2 You said:
Does the shipment mode influence the likelihood of meeting the scheduled delivery date?

````python
# Chart - 2 visualization code
# Create the shipment mode summary
shipment_mode_summary = fd.groupby('Shipment Mode').agg(
    Total_Shipments=('On Time Delivery', 'count'),
    On_Time_Shipments=('On Time Delivery', 'sum')
)

# Calculate on-time delivery rate as a percentage
shipment_mode_summary['On_Time_Delivery_Rate (%)'] = (
    shipment_mode_summary['On_Time_Shipments'] / shipment_mode_summary['Total_Shipments']
) * 100

# Display the summary
shipment_mode_summary = shipment_mode_summary.sort_values(by='On_Time_Delivery_Rate (%)', ascending=False)
shipment_mode_summary
````

````python
# Sort shipment mode summary for plotting
shipment_mode_sorted = shipment_mode_summary.sort_values(by='On_Time_Delivery_Rate (%)', ascending=False)

# Plot bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(shipment_mode_sorted.index, shipment_mode_sorted['On_Time_Delivery_Rate (%)'], color='lightgreen', edgecolor='black')

# Add percentage labels above each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 1, f"{height:.1f}%", ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.title('On-Time Delivery Rate by Shipment Mode')
plt.ylabel('On-Time Delivery Rate (%)')
plt.ylim(0, 110)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

This chart was chosen for its clarity in comparing categorical performance. It clearly compares on-delivery performance across diff. shipment modes.

##### 2. What is/are the insight(s) found from the chart?

According to chart , category "No Mode" have highest on-time rate but can't say if it's most efficient shipping method as it could be non-standard or a data artifact. Air & Air Charter are fairly efficient followed by Truck & Ocean which are less reliable. Ocean has the lowest on time rate.

#### Chart - 3 : Do shipments from certain countries experience more delays compared to others?

````python
# Chart - 3 visualization code
# Group by 'Country' to analyze performance
country_summary = fd.groupby('Country').agg(
    Total_Shipments=('On Time Delivery', 'count'),
    On_Time_Shipments=('On Time Delivery', 'sum'),
    Avg_Delivery_Delay=('Delivery Delay (days)', 'mean')
)

# Calculate on-time delivery rate as a percentage
country_summary['On_Time_Delivery_Rate (%)'] = (
    country_summary['On_Time_Shipments'] / country_summary['Total_Shipments']
) * 100

# Sort by countries with highest average delay
country_summary_sorted = country_summary.sort_values(by='Avg_Delivery_Delay', ascending=False)

# Display countries with the most delivery delays
country_summary_sorted.head(50)
````

````python
# Data for pie chart
labels = country_summary_sorted.index
sizes = country_summary_sorted['On_Time_Delivery_Rate (%)']

# Create pie chart with percentage labels
plt.figure(figsize=(8, 8))
plt.pie(sizes,
        labels=labels,
        autopct='%1.1f%%',        # Shows values like "85.3%"
        startangle=140,
        colors=plt.cm.Set3.colors,
        wedgeprops={'edgecolor': 'black'})  # Optional: adds border to slices

plt.title('On-Time Delivery Rate by Country', fontsize=14)
plt.axis('equal')  # Keep the pie chart circular
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

This chart was chosen as it gives holistic regional prespective across many countries in one place. Also, it helps show how each slice has proportional representations which helps in comparing delivery performance by geography.

Countries like Dominican Republic, Nigeria, and South Africa stand out with higher shares (3.3%-3.4%) — suggesting more shipments or better on-time rates. Countries like Kenya, Senegal, Uganda, and Rwanda are at the lower end (1.7%–2.0%), which may indicate fewer shipments or more delays. Most countries have between 2.0% to 2.6% share

#### Chart - 5:"Is there a difference in delivery performance (on-time vs. delayed) based on the time between the PO Sent to Vendor Date and the Scheduled Delivery Date?"

````python
# Chart - 5 visualization code
# Calculate Lead Time
fd["Lead Time (days)"] = (fd["Scheduled Delivery Date"] - fd["PO Sent to Vendor Date"]).dt.days

# Determine delivery performance
fd["Delivery Status"] = fd["Delivered to Client Date"] <= fd["Scheduled Delivery Date"]
fd["Delivery Status"] = fd["Delivery Status"].map({True: "On-Time", False: "Delayed"})

# Plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=fd, x="Lead Time (days)", hue="Delivery Status", kde=True, bins=50, palette="Set1", alpha=0.6)

# Add mean lines
plt.axvline(fd[fd["Delivery Status"] == "On-Time"]["Lead Time (days)"].mean(), color='blue', linestyle='--', label='On-Time Mean')
plt.axvline(fd[fd["Delivery Status"] == "Delayed"]["Lead Time (days)"].mean(), color='red', linestyle='--', label='Delayed Mean')

# Add titles and labels
plt.title("Distribution of Lead Time by Delivery Performance")
plt.xlabel("Lead Time (days)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

The histogram was selected to effectively compare the distribution of lead times (the number of days between when a PO is sent to a vendor and its scheduled delivery date) for On-Time vs. Delayed deliveries. It visualizes both frequency and spread, making it easy to:
Observe how lead time is associated with delivery performance.
Identify patterns, central tendency, and outliers.
Compare average lead times across performance categories with mean lines.

##### 2. What is/are the insight(s) found from the chart?

Delayed deliveries tend to have shorter lead times compared to on-time deliveries.
The mean lead time for on-time deliveries (blue dashed line) is higher than that for delayed deliveries (red dashed line).
There's a large cluster of delayed deliveries around shorter lead times (between ~50–120 days), indicating insufficient time for fulfillment may be a major factor.
The tail on the right shows that some deliveries had long lead times but were still delayed—pointing toward other factors like logistics inefficiencies.

#### Chart - 10 - Correlation Heatmap
